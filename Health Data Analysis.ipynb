# Analysis

```python
# These are the "Tableau 20" colors as RGB.    
tableau20 = [(31, 119, 180), (255, 127, 14),   
             (44, 160, 44), (227, 119, 194), (255, 152, 150),    
             (148, 103, 189), (197, 176, 213), (140, 86, 75),    
             (247, 182, 210), (214, 39, 40), (127, 127, 127), (199, 199, 199),    
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]    
  
# Scale the RGB values to the [0, 1] range, which is the format matplotlib accepts.    
for i in range(len(tableau20)):    
    r, g, b = tableau20[i]    
    tableau20[i] = (r / 255., g / 255., b / 255.)
```

### USA health expenditure compared to other highly developed countries

#### Per capita

```python
# Select top 10 countries in terms of spending in 2015
top_10 = health_expenditure_capita.nlargest(10,"2015")['Country']

# Filter dataframe to 10 top countries and transpose
filter_top10 = health_expenditure_capita['Country'].isin(top_10)
health_expenditure_capita_top10 = health_expenditure_capita[filter_top10]
health_expenditure_capita_top10.set_index('Country', inplace=True)
health_expenditure_capita_top10 = health_expenditure_capita_top10.transpose()
health_expenditure_capita_top10.sort_index(ascending=True, inplace=True)

health_expenditure_capita_top10.columns = ['Australia',
                                           'Denmark',
                                           'Germany',
                                           'Ireland',
                                           'Luxembourg',
                                           'Netherlands',
                                           'Norway',
                                           'Sweden',
                                           'Switzerland',
                                           'USA']
health_expenditure_capita_top10
```

```python
# Plot data

plt.figure(figsize=(12, 14))   

ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False)  

ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left() 

plt.yticks(fontsize=14)    
plt.xticks(fontsize=14)

for y in range(2000,10001,2000):    
    plt.plot(range(-1, 16), [y] * len(range(-1, 16)), "--", lw=0.5, color="black", alpha=0.3) 

# Remove grey background
ax.set_facecolor('white')

plt.tick_params(axis='both', which='both', bottom=False, top=False,    
                labelbottom=True, left=False, right=False, labelleft=True)

for rank, column in enumerate(health_expenditure_capita_top10.columns):    
    plt.plot(health_expenditure_capita_top10.index.values,    
           health_expenditure_capita_top10[column].values,    
            lw=2.5, color=tableau20[rank])
    
    y_pos = health_expenditure_capita_top10[column].values[-1] - 100    
    if column == "Switzerland":    
        y_pos += 50    
    elif column == "Sweden":    
        y_pos += 100
    elif column == "Denmark":    
        y_pos -= 50
    elif column == "Australia":    
        y_pos += 200        
    elif column == "Netherlands":    
        y_pos += 170
    elif column == "Ireland":    
        y_pos -= 70   
    elif column == "Germany":    
        y_pos -= 120   
        
    plt.text(15.1, y_pos, column, fontsize=14, color=tableau20[rank])   

plt.text(8, 10500, "Health expenditure per capita in US$ - Top 10 countries", fontsize=17, ha="center")

plt.text(0, 500, "Data source: Current health expenditure (CHE) per capita in US$. Geneva, World Health Organization; 2017.", fontsize=10)    
```

#### As % of GDP

```python
# Select top 10 countries in terms of spending in 2015
top_10_gdp = health_expenditure_gdp.nlargest(10,"2015")['Country']

# # Filter dataframe to 10 top countries and transpose
filter_top10_gdp = health_expenditure_gdp['Country'].isin(top_10_gdp)
health_expenditure_gdp_top10 = health_expenditure_gdp[filter_top10_gdp]
health_expenditure_gdp_top10.set_index('Country', inplace=True)
health_expenditure_gdp_top10 = health_expenditure_gdp_top10.transpose()
health_expenditure_gdp_top10.sort_index(ascending=True, inplace=True)

health_expenditure_gdp_top10.columns = ['Andorra',
                                           'Germany',
                                           'Liberia',
                                           'Maldives',
                                           'Marshall Islands',
                                           'Micronesia',
                                           'Sierra Leone',
                                           'Switzerland',
                                           'Tuvalu',
                                           'USA']
health_expenditure_gdp_top10
```

```python
# Plot data

plt.figure(figsize=(12, 14))    

ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False)  

ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left() 

plt.ylim(0, 30)

plt.yticks(range(0, 30, 3), [str(x) + "%" for x in range(0, 30, 3)],fontsize=14)    
plt.xticks(fontsize=14)

for y in range(3,27,3):    
    plt.plot(range(-1, 16), [y] * len(range(-1, 16)), "--", lw=0.5, color="black", alpha=0.3) 

# Remove grey background
ax.set_facecolor('white')

plt.tick_params(axis='both', which='both', bottom=False, top=False,    
                labelbottom=True, left=False, right=False, labelleft=True)

for rank, column in enumerate(health_expenditure_gdp_top10.columns):    
    plt.plot(health_expenditure_gdp_top10.index.values,    
           health_expenditure_gdp_top10[column].values,    
            lw=2.5, color=tableau20[rank])
    
    y_pos = health_expenditure_gdp_top10[column].values[-1]
    if column == "Germany":    
        y_pos -= 0.4    
    elif column == "Switzerland":    
        y_pos += 0.5
    elif column == "Maldives":    
        y_pos -= 0.1
    elif column == "Micronesia":    
        y_pos += 0.3     
    elif column == "Liberia":    
        y_pos += 0.2
    elif column == "Tuvalu":    
        y_pos -= 0.3   
  
    plt.text(15.1, y_pos, column, fontsize=14, color=tableau20[rank])   

plt.text(8, 30, "Health expenditure as % of gross domestic product (GDP) - Top 10 countries", fontsize=17, ha="center")

plt.text(-1, -2, "Data source: Current health expenditure (CHE) as percentage of gross domestic product (GDP) (%). Geneva, World Health Organization; 2017.", fontsize=10)    
```

### Health workforce density

#### Physicians

```python
# Select physicians density from the last available year
health_workforce_phys_sub = health_workforce.loc[:,['Country','Year','Physicians density (per 1000 population)']]
health_workforce_phys_sub.dropna(inplace=True)
health_workforce_phys_sub = health_workforce_phys_sub.loc[health_workforce_phys_sub.groupby('Country')['Year'].idxmax(), :].reset_index(drop=True)

# Merge country codes and workforce datasets
health_workforce_phys_geo = pd.merge(left=health_workforce_phys_sub,right=country_code, left_on='Country', right_on='Country Name')
health_workforce_phys_geo = health_workforce_phys_geo.drop_duplicates() 

# Subset dataset
health_workforce_phys = health_workforce_phys_geo.loc[:,['Country', 'ADM0_A3', 'Physicians density (per 1000 population)']].copy()

# Merge workforce and geometry
health_workforce_phys_density = gdf.merge(health_workforce_phys, on='ADM0_A3', how='left')
health_workforce_phys_density.sort_values(by=['Physicians density (per 1000 population)'], ascending=False, inplace=True)
health_workforce_phys_density.head(10)
```

```python
# Plot data
ax = health_workforce_phys_density.dropna().plot(column='Physicians density (per 1000 population)', cmap='Blues', figsize=(16, 10), scheme='fisher_jenks', k=7, legend=True)

# Draw lines for countries with no data
health_workforce_phys_density[health_workforce_phys_density.isna().any(axis=1)].plot(ax=ax, color='#f2f2f2', hatch='///')

# Add title, data soure, legend
ax.set_title('Global Distribution of Physicians per 1000 population, Latest available year', fontdict={'fontsize': 20}, loc='center')
ax.annotate('Data source: Global Health Workforce Statistics. Geneva, World Health Organization; 2017.', xy=(0.25, 0.1), size=12, xycoords='figure fraction')
ax.set_axis_off()
ax.get_legend().set_bbox_to_anchor((.12, .4))

# Save as png
# plt.savefig('images/physicians.png')
```

#### Nursing and midwifery personnel

```python
# Select nursing and midwifery personnel density from the last available year
health_workforce_nurse_sub = health_workforce.loc[:,['Country','Year','Nursing and midwifery personnel density (per 1000 population)']]
health_workforce_nurse_sub.dropna(inplace=True)
health_workforce_nurse_sub = health_workforce_nurse_sub.loc[health_workforce_nurse_sub.groupby('Country')['Year'].idxmax(), :].reset_index(drop=True)

# Merge country codes and workforce datasets
health_workforce_nurse_geo = pd.merge(left=health_workforce_nurse_sub,right=country_code, left_on='Country', right_on='Country Name')
health_workforce_nurse_geo = health_workforce_nurse_geo.drop_duplicates() 

# Subset dataset
health_workforce_nurse = health_workforce_nurse_geo.loc[:,['Country', 'ADM0_A3', 'Nursing and midwifery personnel density (per 1000 population)']].copy()

# Merge workforce and geometry
health_workforce_nurse_density = gdf.merge(health_workforce_nurse, on='ADM0_A3', how='left')
health_workforce_nurse_density
health_workforce_nurse_density.sort_values(by=['Nursing and midwifery personnel density (per 1000 population)'], ascending=False, inplace=True)
health_workforce_nurse_density.head(10)
```

```python
# Plot data
ax = health_workforce_nurse_density.dropna().plot(column='Nursing and midwifery personnel density (per 1000 population)', cmap='Blues', figsize=(16, 10), scheme='fisher_jenks', k=7, legend=True)

# Draw lines for countries with no data
health_workforce_nurse_density[health_workforce_nurse_density.isna().any(axis=1)].plot(ax=ax, color='#f2f2f2', hatch='///')

# Add title, data soure, legend
ax.set_title('Global Distribution of Nursing and midwifery personnel per 1000 population, Latest available year', fontdict={'fontsize': 20}, loc='center')
ax.annotate('Data source: Global Health Workforce Statistics. Geneva, World Health Organization; 2017.', xy=(0.1, 0.1), size=12, xycoords='figure fraction')
ax.set_axis_off()
ax.get_legend().set_bbox_to_anchor((.12, .4))

# Save as png
# plt.savefig('images/nurse.png')
```

### Leading causes of death

```python
# Subseting dataframe 
deaths_cause_2016 = deaths_cause.loc[:,['cause2015', 'causename', 'age', 'agegroup', 'dths2016']].copy()
death_cause_sum = deaths_cause_2016.groupby(['causename'])['dths2016'].sum().reset_index()
death_cause_sum.sort_values('dths2016', ascending = False, inplace=True)
death_cause_sorted = death_cause_sum.reset_index().loc[:25,['causename','dths2016']].copy()
death_cause_sorted.head(10)
```

```python
# Figure size
plt.figure(figsize=(12, 9))

# Remove plot frame lines
ax = plt.subplot(111)  
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False)    

# Make axis ticks only show up on the bottom and left of the plot
ax.get_xaxis().tick_bottom()  
ax.get_yaxis().tick_left() 

#  Make axis ticks large enough and rotate x ticks
plt.xticks(fontsize=10)  
plt.yticks(range(len(death_cause_sorted.causename)),death_cause_sorted.causename,fontsize=10)

# Change axis labels and make them large enough
plt.xlabel('Number of deaths', fontsize=16)  
plt.ylabel('', fontsize=16)  

# Remove grey background
ax.set_facecolor('white')

# Remove the tick marks
plt.tick_params(axis='both', which='both', bottom=True, top=False,    
                labelbottom=True, left=False, right=False, labelleft=True)
 
# Plot data
death_bar = plt.barh(range(len(death_cause_sorted.causename)),death_cause_sorted.dths2016, color='#3F5D7D', align='center')

# Add title
plt.title("Leading Causes of Death in Females United States, 2016", fontsize=18)  

# Add data source(s)
plt.xlabel("\nData source: Global Health Estimates 2016: Deaths by Cause, Age, Sex, by Country and by Region, 2000-2016. Geneva, World Health Organization; 2018.", fontsize=8)  

death_bar[0].set_color(tableau20[9])
death_bar[1].set_color(tableau20[1])
death_bar[10].set_color(tableau20[1])

plt.show()

# Save as png
plt.savefig('images/causes_death.png')
```
